**Task 2, Question 1**: Clean **all** numerical features so that they can be used in training algorithms. For instance, back_legroom feature is in object format containing both numerical values and text. Extract numerical values (equivalently eliminate the text) so that the numerical values can be used as a regular feature.  
(2 marks)

```python
## Task 2, Question 1 Code Here

import pandas as pd

# Load the test and train datasets
test_df = pd.read_csv('test.csv')
train_df = pd.read_csv('train.csv')

# Display basic information to confirm loading
print("Test Dataset Information:")
print(test_df.info())

print("\nTrain Dataset Information:")
print(train_df.info())
```

```python
# Remove text and get values for back_legroom, front_legroom, maximum_seating, height, wheelbase, width columns
def clean_numeric_columns(df, columns):
    for col in columns:
        # Convert to string, extract numeric parts, handle NaN gracefully
        df[col] = df[col].astype(str).str.extract(r'(\d+\.\d+|\d+)')[0].astype(float)
    return df

# List of columns that need cleaning
columns_to_clean_test = ['back_legroom', 'front_legroom','maximum_seating', 'height', 'wheelbase', 'width']
columns_to_clean_train = ['back_legroom', 'front_legroom','maximum_seating', 'height', 'wheelbase', 'width']

# Clean test dataset
test_df = clean_numeric_columns(test_df, columns_to_clean_test)

# Clean train dataset
train_df = clean_numeric_columns(train_df, columns_to_clean_train)
```

**Task 2, Question 2** Create at least 5 new features from the existing numerical variables which contain multiple items of information, for example you could extract maximum torque and torque rpm from the torque variable.  
(2 marks)

```python
## Task 2, Question 2 Code Here

# Get values from power and torque columns, seperated them into 4 new columns named: power_hp, power_rpm, torque_lbft, torque_rpm
def extract_power_torque_features(df):
    # Extract horsepower as the first number before "hp"
    df['power_hp'] = df['power'].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
    # Extract RPM from power column as the number after "@" symbol
    df['power_rpm'] = df['power'].astype(str).str.extract(r'@ (\d+\.?\d*)')[0].astype(float)
    
    # Extract torque in lb-ft as the first number before "lb-ft"
    df['torque_lbft'] = df['torque'].astype(str).str.extract(r'(\d+\.?\d*)')[0].astype(float)
    # Extract RPM from torque column as the number after "@" symbol
    df['torque_rpm'] = df['torque'].astype(str).str.extract(r'@ (\d+\.?\d*)')[0].astype(float)

    df = df.drop(columns=['power', 'torque'])
    
    return df

test_df = extract_power_torque_features(test_df)
train_df = extract_power_torque_features(train_df)
```

```python
# Create fuel_economy_difference as the difference between highway and city fuel economy
test_df['fuel_economy_difference'] = test_df['highway_fuel_economy'] - test_df['city_fuel_economy']
train_df['fuel_economy_difference'] = train_df['highway_fuel_economy'] - train_df['city_fuel_economy']
```

**Task 2, Question 3**: Impute the missing values for all features in both the training and test datasets.   
(2 marks)

```python
## Task 2, Question 3 Code Here

# Function to check missing values and unique values for each column
def missing_and_unique_values(df):
    missing_values = df.isna().sum()  # Count missing values for each column
    unique_values = df.nunique()      # Count unique values for each column
    summary = pd.DataFrame({'missing_values': missing_values, 'unique_values': unique_values})
    return summary

# Check missing and unique values for the test dataset
print("Test Dataset Missing and Unique Values:")
print(missing_and_unique_values(test_df))

# Check missing and unique values for the train dataset
print("\nTrain Dataset Missing and Unique Values:")
print(missing_and_unique_values(train_df))
```

```python
# Function to impute missing values for numerical and categorical columns

def impute_missing_values(df):
    # Impute numerical columns with mean
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        df[col] = df[col].fillna(df[col].mean())
        
    # Impute categorical columns with mode
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].fillna(df[col].mode()[0])
        
    return df

# Apply imputation to both datasets
test_df = impute_missing_values(test_df)
train_df = impute_missing_values(train_df)
```

```python
# Check missing and unique values for the test dataset
print("Test Dataset Missing and Unique Values:")
print(missing_and_unique_values(test_df))

# Check missing and unique values for the train dataset
print("\nTrain Dataset Missing and Unique Values:")
print(missing_and_unique_values(train_df))
```

**Task 2, Question 4**: Encode all categorical variables appropriately as discussed in class. 

- Where multiple values are given for an observation encode the observation as 'other'. 
- Where a categorical feature contains more than 5 unique values, map the features into 5 most frequent values + 'other' and then encode appropriately. For instance, if appropriate, you could all map colours into {red, yellow, green, blue, purple, other} and then encode.  
(2 marks)

```python
## Task 2, Question 4 Code Here

# Identify categorical columns
categorical_columns_test = test_df.select_dtypes(include=['object']).columns.drop('vin')
categorical_columns_train = train_df.select_dtypes(include=['object']).columns.drop('vin')
```

```python
# Step 1: Replace multiple categories with 'other'
def replace_multiple_categories(df, columns):
    for col in columns:
        # Replace entries with multiple categories with 'other'
        df[col] = df[col].apply(lambda x: 'Other' if '/' in str(x) else x)
    return df
```

```python
# Apply replacement to both datasets
test_df = replace_multiple_categories(test_df, categorical_columns_test)
train_df = replace_multiple_categories(train_df, categorical_columns_train)
```

```python
# Function to limit categories and display the top 5 most frequent values for each categorical column
def limit_categories_with_display(df, columns, top_n=5):
    top_frequent_values = {}
    for col in columns:
        # Get the top N most frequent values and store them
        top_values = df[col].value_counts().nlargest(top_n).index
        top_frequent_values[col] = top_values.tolist()
        
        # Map all values not in the top N to 'other'
        df[col] = df[col].apply(lambda x: x if x in top_values else 'Other')
    return df, top_frequent_values

# Apply and get top 5 frequent values for each categorical column in test and train datasets
test_df, top_frequent_values_test = limit_categories_with_display(test_df, categorical_columns_test)
train_df, top_frequent_values_train = limit_categories_with_display(train_df, categorical_columns_train)
```

```python
# Function to display unique values for each categorical column in a dataset
def unique_values_for_categorical(df):
    categorical_columns = df.select_dtypes(include=['object']).columns.drop('vin')
    unique_values = {col: df[col].unique() for col in categorical_columns}
    return unique_values

# Get unique values for each categorical column in the test dataset
unique_values_test = unique_values_for_categorical(test_df)
print("Unique Values for Each Categorical Column in Test Dataset:")
for col, values in unique_values_test.items():
    print(f"{col}: {values}")

# Get unique values for each categorical column in the train dataset
unique_values_train = unique_values_for_categorical(train_df)
print("\nUnique Values for Each Categorical Column in Train Dataset:")
for col, values in unique_values_train.items():
    print(f"{col}: {values}")
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder to create the mappings
label_encoder = LabelEncoder()

# Function to create a mapping dictionary for each categorical column
def create_encoding_map(df, columns):
    encoding_map = {}
    for col in columns:
        # Fit the encoder on the column and generate a dictionary of mappings
        label_encoder.fit(df[col].astype(str))
        encoding_map[col] = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    return encoding_map

# Generate mappings for both the test and train datasets (using the same columns for consistency)
encoding_map_test = create_encoding_map(test_df, categorical_columns_test)
encoding_map_train = create_encoding_map(train_df, categorical_columns_train)

# Display the mappings for categorical columns in test and train datasets
print("Encoding Mappings for Categorical Columns in Test Dataset:")
for col, mapping in encoding_map_test.items():
    print(f"{col}: {mapping}")

print("\nEncoding Mappings for Categorical Columns in Train Dataset:")
for col, mapping in encoding_map_train.items():
    print(f"{col}: {mapping}")
```
